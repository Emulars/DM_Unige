```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph from exercise1.py
G = nx.read_adjlist("exercise1_graph.adjlist")

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Compute betweenness centrality and print it
betweenness_centrality = nx.betweenness_centrality(G)

# Compute closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Print top 10 nodes with their betweenness centrality
print("Top 10 nodes by betweenness centrality:")
for node, centrality in sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"Node {node}: {centrality}")

# Print a separator
print("-" * 50)

# Print top 10 nodes with their closeness centrality
print("Top 10 nodes by closeness centrality:")
for node, centrality in sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"Node {node}: {centrality}")
```
