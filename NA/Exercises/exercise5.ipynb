```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Creaate a complete graph
G_complete = nx.complete_graph(10)

print("Complete Graph")

# Print the number of nodes
print("Number of nodes:", G_complete.number_of_nodes())

# Print the number of edges
print("Number of edges:", G_complete.number_of_edges())

# Print the min degree
print("Minimum degree:", min(dict(G_complete.degree()).values()))

# Print the max degree
print("Maximum degree:", max(dict(G_complete.degree()).values()))

# Print the diameter
print("Diameter:", nx.diameter(G_complete))

# Print the number of triangles
print("Number of triangles:", sum(nx.triangles(G_complete).values()) / 3)

# Print the transitivity
print("Transitivity:", nx.transitivity(G_complete))

# Print the clustering coefficient of the entire network
print("Clustering coefficient:", nx.average_clustering(G_complete))

# Print the average clustering coefficient
print("Average clustering coefficient:", nx.average_clustering(G_complete))

# Plot the graph and the its degree distribution toghether
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G_complete, with_labels=True, node_size=400, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", font_family="sans-serif", edge_color="gray", pos=nx.circular_layout(G_complete))
plt.subplot(122)
plt.hist(dict(G_complete.degree()).values(), bins=range(0, 11), rwidth=0.8, color="skyblue", edgecolor="black")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution")

plt.tight_layout()
plt.show()
```

```python
def force_directed_layout(G, iterations):
    # Initialize node positions
    pos = nx.spring_layout(G)

    # Draw initial layout
    nx.draw(G, pos, with_labels=True)
    plt.title("Initial Layout")
    plt.show()

    # Perform force-directed layout algorithm
    for _ in range(iterations):
        # Compute forces
        pos = nx.spring_layout(G, pos=pos, iterations=1)

    # Draw final layout
    nx.draw(G, pos, with_labels=True)
    plt.title("Final Layout")
    plt.show()

    return pos

# Create a random graph
# G = nx.erdos_renyi_graph(10, 0.15)  # Create a random graph with 10 nodes and edge probability 0.15
G = nx.gnm_random_graph(10, 15)
pos = force_directed_layout(G, iterations=100)
print("Final node positions:", pos)
```

```python
# Create a Watts-Strogatz model
G_watts = nx.watts_strogatz_graph(10, 4, 0.3)

print("\nWatts-Strogatz Model")

# Print the number of nodes
print("Number of nodes:", G_watts.number_of_nodes())

# Print the number of edges
print("Number of edges:", G_watts.number_of_edges())

# Print the min degree
print("Minimum degree:", min(dict(G_watts.degree()).values()))

# Print the max degree
print("Maximum degree:", max(dict(G_watts.degree()).values()))

# Print the diameter
print("Diameter:", nx.diameter(G_watts))

# Print the number of triangles
print("Number of triangles:", sum(nx.triangles(G_watts).values()) / 3)

# Print the transitivity
print("Transitivity:", nx.transitivity(G_watts))

# Print the clustering coefficient of the entire network
print("Clustering coefficient:", nx.average_clustering(G_watts))

# Print the average clustering coefficient
print("Average clustering coefficient:", nx.average_clustering(G_watts))

# Generate the circle layout for the Watts-Strogatz model
pos_watts = nx.circular_layout(G_watts)

# Plot the graph and its degree distribution together
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G_watts, with_labels=True, node_size=400, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", font_family="sans-serif", edge_color="gray", pos=pos_watts)
plt.subplot(122)
plt.hist(dict(G_watts.degree()).values(), bins=range(0, 11), rwidth=0.8, color="skyblue", edgecolor="black")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Distribution")

plt.tight_layout()
plt.show()
```

```python
# Set the number of nodes and edges
N = 10
M = 20

# Generate a Watts-Strogatz graph
G_watts = nx.watts_strogatz_graph(N, 4, 0.3)

# Generate a G(N,M) random graph
G_random = nx.gnm_random_graph(N, M)

# Compare the characteristics of the two graphs
print("Watts-Strogatz Model")
print("Number of nodes:", G_watts.number_of_nodes())
print("Number of edges:", G_watts.number_of_edges())
print("Minimum degree:", min(dict(G_watts.degree()).values()))
print("Maximum degree:", max(dict(G_watts.degree()).values()))
print("Diameter:", nx.diameter(G_watts))
print("Number of triangles:", sum(nx.triangles(G_watts).values()) / 3)
print("Transitivity:", nx.transitivity(G_watts))
print("Clustering coefficient:", nx.average_clustering(G_watts))

print("\nG(N,M) Random Graph")
print("Number of nodes:", G_random.number_of_nodes())
print("Number of edges:", G_random.number_of_edges())
print("Minimum degree:", min(dict(G_random.degree()).values()))
print("Maximum degree:", max(dict(G_random.degree()).values()))
print("Diameter:", nx.diameter(G_random))
print("Number of triangles:", sum(nx.triangles(G_random).values()) / 3)
print("Transitivity:", nx.transitivity(G_random))
print("Clustering coefficient:", nx.average_clustering(G_random))

# Plot the two graphs
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G_watts, with_labels=True, node_size=400, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", font_family="sans-serif", edge_color="gray")
plt.title("Watts-Strogatz Model")
plt.subplot(122)
nx.draw(G_random, with_labels=True, node_size=400, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", font_family="sans-serif", edge_color="gray")
plt.title("G(N,M) Random Graph")

plt.tight_layout()
plt.show()
```
